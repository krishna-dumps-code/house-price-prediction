```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
house = pd.read_csv('house_b4_encoding.csv')
house.head()
```

# Encoding of Categorical features
Encoding of ordinal and nominal features with Label Encoding and One Hot Encoding respectively.

```python
categorical_feat = [feature for feature in house.columns if house[feature].dtype=='O']
for feat in categorical_feat:
    print(feat,': cardinality = ',len(house[feat].unique()))
```

```python
for feat in categorical_feat:
    sns.catplot(y='SalePrice',x=feat,data=house.sort_values('SalePrice',ascending=False),kind='boxen',height=5,aspect=2)
    plt.show()
```

Features like (GarageFinish, KitchenQual, CentralAir, BsmtQual, ExterQual) can be Label Encoded and others can be Ohe Hot Encoded.

```python
label_feat = ['GarageFinish','KitchenQual','CentralAir','BsmtQual','ExterQual']
OHE_feat = [feat for feat in categorical_feat if feat not in label_feat]
```

```python
house1 = pd.get_dummies(house[OHE_feat],drop_first=True)
house1.head()
```

```python
house.drop(OHE_feat,axis=1,inplace=True)
```

```python
house.drop('Id',axis=1,inplace=True)
```

```python
house = pd.concat([house,house1],axis=1)
house.head()
```

```python
house.shape
```

```python
#Target Label Encoding
house['KitchenQual'].replace({'Ex':3,'Gd':2,'TA':1,'Fa':0},inplace=True)
house['KitchenQual'].unique()
```

```python
house['BsmtQual'].replace({'Ex':4,'Gd':3,'TA':2,'Fa':1,'Missing':0},inplace=True)
house['CentralAir'].replace({'Y':1,'N':0},inplace=True)
house['GarageFinish'].replace({'Fin':3,'RFn':2,'Unf':1,'Missing':0},inplace=True)
house['ExterQual'].replace({'Ex':3,'Gd':2,'TA':1,'Rare':0},inplace=True)
```

```python
house.head()
```

```python
house_scaled = house.copy()
from sklearn.preprocessing import MinMaxScaler
feat_scale = [feature for feature in house_scaled.columns if feature != 'SalePrice']
scaler = MinMaxScaler()
scaler.fit(house_scaled[feat_scale])
```

```python
house_scaled1 = pd.concat([house_scaled['SalePrice'].reset_index(drop=True),
                   pd.DataFrame(scaler.transform(house_scaled[feat_scale]),columns=feat_scale)],axis=1)
house_scaled1.head()
```

```python
y = house_scaled1['SalePrice']
X =house_scaled1.drop(['SalePrice'],axis=1)
```

```python
#Checking Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feat_select_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feat_select_model.fit(X,y)
```

```python
selected_feat = X.columns[(feat_select_model.get_support())]
print('Total features:',X.shape[1])
print('Selected features:',len(selected_feat))
```

```python
house2 = pd.concat([X[selected_feat],y],axis=1)
house2.head()
```

```python
house2.to_csv('House_ordinal_nominal.csv',index=False)
```

```python
plt.figure(figsize=(20,15))
corr = house2.corr()
sns.heatmap(corr,annot=True,cmap=sns.diverging_palette(230, 20, as_cmap=True))
```
